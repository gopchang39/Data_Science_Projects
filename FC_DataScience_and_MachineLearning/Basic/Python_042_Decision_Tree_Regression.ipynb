# Decision Tree Regression Basic

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
# Load data
data = np.sort(np.random.uniform(low=0, high=5, size=(80,1)))
label = np.sin(data).ravel()
label[::5] += 3 * (0.5 - np.random.uniform(0, 1, 16))
plt.figure(figsize=(8,8))
plt.scatter(data, label)
```

```python
# visualization
viz_test_data = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
viz_test_data[:5]
```

## Tree depth study

```python
# one depth
from sklearn.tree import DecisionTreeRegressor, plot_tree
first_divide = DecisionTreeRegressor(max_depth=1)
first_divide.fit(data, label)
plot_tree(first_divide)

firt_divide_predictions = first_divide.predict(viz_test_data)
plt.figure(figsize=(8,8))
plt.axvline(first_divide.tree_.threshold[0], color="red")
plt.scatter(data,label)
plt.plot(viz_test_data, firt_divide_predictions, color='C2', linewidth=2)
```

```python
# two depths
from sklearn.tree import DecisionTreeRegressor, plot_tree
first_divide = DecisionTreeRegressor(max_depth=2)
first_divide.fit(data, label)
plot_tree(first_divide)

firt_divide_predictions = first_divide.predict(viz_test_data)
plt.figure(figsize=(8,8))
plt.axvline(first_divide.tree_.threshold[0], color="red")
plt.scatter(data,label)
plt.plot(viz_test_data, firt_divide_predictions, color='C2', linewidth=2)
```

```python
# five depths
from sklearn.tree import DecisionTreeRegressor, plot_tree
first_divide = DecisionTreeRegressor(max_depth=5)
first_divide.fit(data, label)
plot_tree(first_divide)

firt_divide_predictions = first_divide.predict(viz_test_data)
plt.figure(figsize=(8,8))
plt.axvline(first_divide.tree_.threshold[0], color="red")
plt.scatter(data,label)
plt.plot(viz_test_data, firt_divide_predictions, color='C2', linewidth=2)
```
