Q1. Print out the name and address of the campling sites.

```python
%%sql
select 사업장명 as Name, 소재지전체주소 as address from camping_info ;
```

Q2. Print the sites which are open for buisiness.

```python
%%sql
select 사업장명 as Name, 소재지전체주소 as address from camping_info where 영업상태구분코드 = 1 ;
```

Q3. Camping sites located in 양양.

```python
%%sql
select 사업장명 as Name, 소재지전체주소 as address from camping_info where 소재지전체주소 like "%양양%" ;
```

Q4. Closed sites from Q3.

```python
%%sql
select 사업장명 as Name, 소재지전체주소 as address from camping_info where 소재지전체주소 like "%양양%" and 영업상태구분코드 = 3 ;
```

Q5. Camping sites in 태안.

```python
%%sql
select 사업장명 as Name, 소재지전체주소 as address from camping_info where 소재지전체주소 like "%태안%";
```

Q6. From Q5, list the sites which are closed in 2020.

```python
%%sql
select 사업장명 as Name, 소재지전체주소 as address, 폐업일자 as closure_date from camping_info where 소재지전체주소 like "%태안%" and 폐업일자 like "%2020%" ;
```

Q7. List of Camping sites in 서울 and 제주도.

```python
%%sql
select 사업장명 as Name, 소재지전체주소 as address from camping_info where (소재지전체주소 like "%서울%" or 소재지전체주소 like "%제주도%") ;
```
